```python
jobs['year'] = pd.DatetimeIndex(jobs['post_date'], yearfirst = True).year
```

```python
annual_stats = jobs.groupby(['year','state']).mean().reset_index()
annual_stats2 = annual_stats.loc[annual_stats['state'].isin(['NY', 'CA', 'TX'])].round({'salary':2})
annual_stats3 = annual_stats2.pivot_table(index = 'state', columns = 'year', values = 'salary')
annual_stats3
```
